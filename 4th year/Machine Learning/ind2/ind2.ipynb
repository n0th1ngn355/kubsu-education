```python
column_names = ['Elevation',
 'Aspect',
 'Slope',
 'Horizontal_Distance_To_Hydrology',
 'Vertical_Distance_To_Hydrology',
 'Horizontal_Distance_To_Roadways',
 'Hillshade_9am',
 'Hillshade_Noon',
 'Hillshade_3pm',
 'Horizontal_Distance_To_Fire_Points',
 'Wilderness_Area_0',
 'Wilderness_Area_1',
 'Wilderness_Area_2',
 'Wilderness_Area_3'] + [f'Soil_Type_{i}' for i in range(40)] + ['Cover_Type']
```

```python
import pandas as pd


df = pd.read_csv('covertype/covtype.data', header=None, names=column_names)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df.hist(figsize=(18, 16))
plt.show()
```

```python


X, y = df.drop(['Cover_Type'], axis=1), df[['Cover_Type']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

cont_columns = column_names[:10]
mscaler = MinMaxScaler()
X_train = mscaler.fit_transform(X_train)
X_test = mscaler.transform(X_test)
```

```python
sns.heatmap(df[cont_columns + ['Cover_Type']].corr())
plt.show()
```

```python
# Оценка важности признаков

from sklearn.feature_selection import mutual_info_classif
mutual_info = mutual_info_classif(X, y)
mutual_info = pd.Series(mutual_info)
mutual_info.index = X.columns
mutual_info.sort_values(ascending=False)
```

```python
# Выделяем 20 наиболее важных

from sklearn.feature_selection import SelectKBest
sel_cols = SelectKBest(mutual_info_classif, k=20)
sel_cols.fit(X_train, y_train)
col = X_train.columns[sel_cols.get_support()]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier


log_model = LogisticRegression()
rf = RandomForestClassifier()

log_model.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

knn = KNeighborsClassifier()

knn.fit(X_train, y_train)

# Зададим сетку - среди каких значений выбирать наилучший параметр.
# knn_grid = {'n_neighbors': np.arange(10, 100, 10)}

# gs = GridSearchCV(knn, knn_grid, cv=3)

# gs.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score

def metrics_report(y_true, y_pred):
    print(f'precision: {precision_score(y_true, y_pred, average='macro')}')
    print(f'recall: {recall_score(y_true, y_pred, average='macro')}')
    print(f'f1: {f1_score(y_true, y_pred, average='macro')}')
```

```python
metrics_report(y_test, log_model.predict(X_test))
print()
metrics_report(y_test, rf.predict(X_test))
print()
metrics_report(y_test, knn.predict(X_test))
```
