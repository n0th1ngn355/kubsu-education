```python
df['Genre'] = df['Genre'].apply(lambda x: int(x=='Male'))
```

```python
import matplotlib.pyplot as plt

data = df['Genre'].value_counts()
plt.bar(data.keys(), data.values)
plt.show()
```

```python
plt.pie(df['Genre'].value_counts(), labels=['Female', 'Male'], autopct='%1.1f%%')
plt.show()
```

```python
plt.hist(df['Age'])
plt.show()
```

```python
plt.boxplot(df['Age'])
plt.show()
```

```python
plt.hist(df['Annual Income (k$)'])
plt.show()
```

```python
df['Annual Income (k$)'].plot(kind='density')
plt.show()
```

```python
df['Spending Score (1-100)'].hist()
```

```python
data = df['Spending Score (1-100)'].value_counts()
plt.pie(data.keys(), data.values, autopct='%1.1f%%')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
y_pred = KMeans(n_clusters=5, random_state=1, n_init='auto').fit_predict(df)
plt.scatter(list(range(df['Spending Score (1-100)'].shape[0])), df['Spending Score (1-100)'], c=y_pred)
plt.title("  ")
plt.show()
```
