```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import numpy as np

f = lambda x: x*x - 2*x + 3
e = 0.2
l = 0.5

style = "Simple, tail_width=0.5, head_width=4, head_length=8"
kw = dict(arrowstyle=style, color="k")

def draw(x, a, b, label):
    X = np.linspace(-2, 8, 100)
    plt.figure(figsize=(16,9))
    plt.plot(X, f(X))
    plt.scatter(x, f(x))
    for i in range(len(a)-1):
        t1, t2, ind = a[i][0], a[i+1][0], a[i+1][1]
        ar = patches.FancyArrowPatch((t1, f(t1)), (t2, f(t2)), connectionstyle="arc3,rad=-.5", **kw)
        plt.text(t1+(t2-t1)/2, 1+f(t1)+(f(t2)-f(t1))/2, f"{ind}")
        plt.gca().add_patch(ar)
    for i in range(len(b)-1):
        t1, t2, ind = b[i][0], b[i+1][0], b[i+1][1]
        ar = patches.FancyArrowPatch((t1, f(t1)), (t2, f(t2)), connectionstyle="arc3,rad=.5", **kw)
        plt.text(t1+(t2-t1)/2, 1 + f(t1)+(f(t2)-f(t1))/2, f"{ind}")
        plt.gca().add_patch(ar)
    t1, t2 = a[-1][0], b[-1][0]
    plt.vlines(t1, f(t1), 10, 'g', label=f"a: {t1:.6f}")
    plt.vlines(t2, f(t2), 10, 'g', label=f"b: {t2:.6f}")
    plt.text(x, f(x)-1.4, f"{x:.6f}")
    plt.legend()
    plt.title(label)
    plt.show()

# f = lambda x: x*x + 6*x + 13
```

```python
a,b = -2, 8
i = 0
changes = {'a':[(a,0)], 'b': [(b,0)]}

while b-a > l:
    print(f"Итерация {i}: [{a:.3f}, {b:.3f}]")
    y = (a+b-e)/2
    z = (a+b+e)/2
    print(f'y{i} = ({a:.3f}+{b:.3f}-{e})/2 = {y:.3f}')
    print(f'z{i} = ({a:.3f}+{b:.3f}+{e})/2 = {z:.3f}')
    f1, f2 = f(y), f(z)
    print(f'f(y{i}) = {f1:.3f}')
    print(f'f(z{i}) = {f2:.3f}')
    if f1 <= f2:
        b = z
        changes['b'].append((b, i))
    else:
        a = y
        changes['a'].append((a, i))
    i+=1
x = (a+b)/2
print('Точка:',x)
print('Значение функции:',f(x))
N = i*2
print('N:', N)
print('R(N) =', 1/(2**(N//2)))

draw(x, changes['a'], changes['b'], 'Метод дихотомии')
```

```python
a,b = -2, 8
i = 0
sech = (3-5**(0.5)) / 2
y = a + sech*(b-a)
z = a+b-y
changes = {'a':[(a,0)], 'b': [(b,0)]}

while b-a > l:
    print(f"Итерация {i}: [{a:.3f}, {b:.3f}]")
    print(f'y{i} = {y:.3f}')
    print(f'z{i} = {z:.3f}')
    f1, f2 = f(y), f(z)
    print(f'f(y{i}) = {f1:.3f}')
    print(f'f(z{i}) = {f2:.3f}')
    if f1 <= f2:
        b = z
        z = y
        y = a+b-y
        changes['b'].append((b, i))
    else:
        a = y
        y = z
        z = a+b-z
        changes['a'].append((a, i))
    i+=1
x = (a+b)/2
print(f'Точка: {a}+{b}/2 = ',x)
print('Значение функции:',f(x))
N = i+1
print('N:', N)
print('R(N) =', 0.618**(N-1))

draw(x, changes['a'], changes['b'], 'Метод золотого сечения')
```

```python
a, b = -2, 8

F = [1, 1]

while F[-1] <= (b-a)/l:
    F.append(F[-1] + F[-2])

n = len(F)

y = a + (b - a)*F[n-3]/F[n-1]
z = a + (b - a)*F[n-2]/F[n-1]
f1 = f(y)
f2 = f(z)

changes = {'a':[(a,0)], 'b': [(b,0)]}
i = 0

for k in range(1, n):
    print(f"Итерация {i}: [{a:.3f}, {b:.3f}]")
    print(f'y{i} = {y:.3f}')
    print(f'z{i} = {z:.3f}')
    print(f'f(y{i}) = {f1:.3f}')
    print(f'f(z{i}) = {f2:.3f}')
    if f1 > f2:
        a = y
        y = z
        f1 = f2
        z = a + F[n-k-1] / F[n-k] * (b - a)
        f2 = f(z)
        changes['a'].append((a, i))
    else:
        b = z
        z = y
        f2 = f1
        y = a + F[n-k-2] / F[n-k] * (b - a)
        f1 = f(y)
        changes['b'].append((b, i))
    i+=1

x = (a+b)/2
print(f'Точка: {a}+{b}/2 = ',x)
print('Значение функции:',f(x))
N = n
print('N:', N)
print('R(N) =', (b-a)/(10))

draw(x, changes['a'], changes['b'], 'Метод Фибоначчи')

```
