```python
import numpy as np
import random as rd
import pandas as pd

np.random.seed(0)
rd.seed(0)
X_train = np.random.rand(20, 2)
y_train = np.where(X_train[:, 0] > X_train[:, 1], 1, -1)

X_test = np.random.rand(1000, 2)
y_test = np.where(X_test[:, 0] > X_test[:, 1], 1, -1)

dfTrain = pd.DataFrame(X_train, columns=['x1', 'x2'])
dfTrain['y'] = y_train
dfTrain
```

```python
nu = 0.4

class Perceptron:
    def __init__(self):
        self.w = np.array([rd.random(), rd.random()])
        self.bias = rd.random()
        leng = (sum(map(lambda x: x*x, self.w)) + self.bias ** 2) ** 0.5
        self.w /= leng
        self.bias /= leng
        self.n_iters = 1

    def fit(self, X, y):
        for _ in range(self.n_iters):
            for i, x_i in enumerate(X):
                y_predicted = self.predict(x_i)
                
                update = nu * (y[i] - y_predicted)
                self.w += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.w) + self.bias
        return np.where(linear_output > 0, 1, -1)

```

```python
nn = Perceptron()
nn.fit(X_train, y_train)

y_pred = nn.predict(X_test)

accuracy = np.mean(y_pred == y_test)

print(f"Точность перцептрона на тестовых данных: {accuracy:.2f}")
```

```python
nu = 0.2

class Adaline:
    def __init__(self):
        self.w = np.array([rd.random(), rd.random()])
        self.bias = rd.random()
        leng = (sum(map(lambda x: x*x, self.w)) + self.bias ** 2) ** 0.5
        self.w /= leng
        self.bias /= leng
        self.n_iters = 1

    def fit(self, X, y):
        for _ in range(self.n_iters):
            for i, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.w) + self.bias

                update = nu * (y[i] - linear_output)
                self.w += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.w) + self.bias
        y_predicted = np.where(linear_output > 0, 1, -1)
        return y_predicted
```

```python
adNN = Adaline()
adNN.fit(X_train, y_train)

y_pred2 = adNN.predict(X_test)

accuracy = np.mean(y_pred2 == y_test)

print(f"Точность адалайна на тестовых данных: {accuracy:.2f}")
```
